```python
from pysus.online_data.SIH import SIH

sih = SIH().load()

files = sih.get_files("RD", uf="RR", year=2025, month=12)
print(files, len(files))

parquets = sih.download(files)

df = parquets.to_dataframe()   # <- aqui está a correção

print(df.shape)
df.head()
```

```python
# =========================
# 1. VISÃO GERAL
# =========================
print("Shape:", df.shape)
display(df.head(3))

# =========================
# 2. TIPOS E COLUNAS
# =========================
print("\nTipos de variáveis:")
display(df.dtypes)

print("\nColunas:")
print(list(df.columns))

# =========================
# 3. MISSING VALUES
# =========================
missing = df.isna().sum().sort_values(ascending=False)
print("\nTop 20 colunas com mais NA:")
display(missing.head(20))

# =========================
# 4. DUPLICATAS
# =========================
dup = df.duplicated().sum()
print(f"\nDuplicatas: {dup}")

# =========================
# 5. LIMPEZA INICIAL
# =========================
df.columns = df.columns.str.lower()

# exemplo: tratar idade inválida (se existir)
if "idade" in df.columns:
    df["idade"] = df["idade"].replace({999: None})

# =========================
# 6. VARIÁVEIS CHAVE (AJUSTE CONFORME COLUNAS)
# =========================
variaveis_interesse = [col for col in ["sexo", "idade", "municip_res", "diag_princ"] if col in df.columns]

print("\nVariáveis disponíveis para análise:")
print(variaveis_interesse)

# =========================
# 7. DESCRITIVA BÁSICA
# =========================
display(df[variaveis_interesse].describe(include="all"))

# =========================
# 8. FREQUÊNCIA (EXEMPLO)
# =========================
if "sexo" in df.columns:
    print("\nDistribuição por sexo:")
    display(df["sexo"].value_counts(dropna=False))

# =========================
# 9. GRÁFICO SIMPLES
# =========================
import matplotlib.pyplot as plt

if "idade" in df.columns:
    df["idade"].dropna().hist()
    plt.title("Distribuição de Idade")
    plt.xlabel("Idade")
    plt.ylabel("Frequência")
    plt.show()

# =========================
# 10. SALVAR DADOS LIMPOS
# =========================
df.to_parquet("sih_rr_2025_12_clean.parquet", index=False)
```

```python
# =========================
# CONFIGURAÇÃO PADRÃO PUBLICAÇÃO
# =========================
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams.update({
    "font.family": "Times New Roman",
    "font.size": 10
})

# =========================
# 1. DISTRIBUIÇÃO DE IDADE (HISTOGRAMA + KDE)
# =========================
if "idade" in df.columns:
    idade = pd.to_numeric(df["idade"], errors="coerce").dropna()

    plt.figure()
    idade.plot(kind="hist", bins=20, density=True)
    idade.plot(kind="kde")
    plt.xlabel("Idade (anos)")
    plt.ylabel("Densidade")
    plt.title("Distribuição de Idade - SIH/SUS (RR, dez/2025)")
    plt.tight_layout()
    plt.savefig("fig_idade.png", dpi=300)
    plt.show()

# =========================
# 2. DISTRIBUIÇÃO POR SEXO (BARRAS)
# =========================
if "sexo" in df.columns:
    sexo_map = {1: "Masculino", 3: "Feminino"}
    sexo = df["sexo"].map(sexo_map)

    freq = sexo.value_counts(normalize=True)*100

    plt.figure()
    freq.sort_values().plot(kind="barh")
    plt.xlabel("Proporção (%)")
    plt.title("Distribuição por Sexo - SIH/SUS (RR, dez/2025)")
    plt.tight_layout()
    plt.savefig("fig_sexo.png", dpi=300)
    plt.show()

# =========================
# 3. TOP 10 DIAGNÓSTICOS (CID-10)
# =========================
if "diag_princ" in df.columns:
    top_diag = df["diag_princ"].value_counts().head(10)

    plt.figure()
    top_diag.sort_values().plot(kind="barh")
    plt.xlabel("Frequência")
    plt.title("Top 10 Diagnósticos Principais (CID-10)")
    plt.tight_layout()
    plt.savefig("fig_diag.png", dpi=300)
    plt.show()
```

```python
# =========================
# CONFIGURAÇÃO PROFISSIONAL
# =========================
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams.update({
    "font.family": "DejaVu Sans",  # compatível com Colab
    "font.size": 11,
    "axes.labelsize": 11,
    "axes.titlesize": 12
})

# =========================
# 1. CRIAR FAIXAS ETÁRIAS
# =========================
df["idade"] = pd.to_numeric(df["idade"], errors="coerce")

bins = [0, 19, 39, 59, 79, 120]
labels = ["0–19", "20–39", "40–59", "60–79", "80+"]

df["faixa_etaria"] = pd.cut(df["idade"], bins=bins, labels=labels, right=False)

# =========================
# 2. PROPORÇÃO POR SEXO + IC95%
# =========================
sexo_map = {1: "Masculino", 3: "Feminino"}
df["sexo_cat"] = df["sexo"].map(sexo_map)

freq = df["sexo_cat"].value_counts()
n = freq.sum()
prop = freq / n

# IC binomial aproximado (normal)
ic_inf = prop - 1.96 * np.sqrt((prop * (1 - prop)) / n)
ic_sup = prop + 1.96 * np.sqrt((prop * (1 - prop)) / n)

# =========================
# GRÁFICO 1 — SEXO COM IC
# =========================
plt.figure()

y = np.arange(len(prop))
plt.barh(y, prop.values)

plt.errorbar(prop.values, y,
             xerr=[prop.values - ic_inf.values, ic_sup.values - prop.values],
             fmt='none')

plt.yticks(y, prop.index)
plt.xlabel("Proporção")
plt.title("Distribuição por Sexo com IC95%")

plt.tight_layout()
plt.savefig("fig_sexo_ic.png", dpi=300)
plt.show()

# =========================
# 3. PIRÂMIDE ETÁRIA (REALMENTE PUBLICÁVEL)
# =========================
pivot = df.groupby(["faixa_etaria", "sexo_cat"]).size().unstack(fill_value=0)

pivot["Masculino"] = -pivot["Masculino"]  # lado esquerdo

plt.figure()

plt.barh(pivot.index, pivot["Masculino"], label="Masculino")
plt.barh(pivot.index, pivot["Feminino"], label="Feminino")

plt.xlabel("Número de internações")
plt.title("Pirâmide Etária das Internações (SIH/SUS - RR, dez/2025)")
plt.legend()

plt.tight_layout()
plt.savefig("fig_piramide.png", dpi=300)
plt.show()

# =========================
# 4. TOP DIAGNÓSTICOS (COM %)
# =========================
top_diag = df["diag_princ"].value_counts(normalize=True).head(10)*100

plt.figure()
top_diag.sort_values().plot(kind="barh")

plt.xlabel("Proporção (%)")
plt.title("Top 10 Diagnósticos Principais (CID-10)")

plt.tight_layout()
plt.savefig("fig_diag_top10.png", dpi=300)
plt.show()
```
